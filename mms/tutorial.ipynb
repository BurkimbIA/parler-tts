```python
from transformers import pipeline
import scipy

model_id = "facebook/mms-tts-mos"
synthesiser = pipeline("text-to-speech", model_id, device=0) # add device=0 if you want to use a GPU

speech = synthesiser("Logtoɛɛmbã vɛɛsg sẽn kẽed ne bã-kãngã kõta vẽenem tɩ bãagã pãng tũuda ne ninsaal fãa ne a yĩnga : neb sãnd yĩns zãad-a lame n kaoos tɩ bũmb ka maan-b ye, la sãnda meng yĩnsã wat n le paama zu-loeesa a taaba wala sãoorã bãagã bi karsinom epatoselilɛɛrã")

scipy.io.wavfile.write("finetuned_output.wav", rate=speech["sampling_rate"], data=speech["audio"][0])
```
